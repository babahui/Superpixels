```python
from numpy import linalg as LA
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.axes3d import Axes3D
    
%matplotlib inline

def plot_eigenspace(mat, label):
        
    # construct desnse distance matrix & eigen-decompostion
    dis_mat = distance_metric(mat)
    w, v = LA.eigh(dis_mat)
    print("v, ", v.shape)
    
#     # eigen-space 3D plot
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(v[:, 0], v[:, 1], v[:, 2], c=label)
    
    
def distance_metric(mat, method='color'):

    if method == 'color':
        if len(mat.shape) == 2:
            print('ok')
            mat = mat.flatten()
            dis_mat = np.zeros((mat.shape[0], mat.shape[0]))

            for i in range(mat.shape[0]):
                for j in range(mat.shape[0]):
                    dis_mat[i, j] = LA.norm(mat[i] - mat[j])
    return dis_mat   
```

```python
# 手工生成, 颜色值相同
mat = np.zeros((20,20))
mat[:10, :10] = np.ones((10, 10))
# plt.imshow(mat)
# plt.show()
label = mat.flatten()
print(label.shape, mat.shape)

plot_eigenspace(mat, label)


# 手工生成, 颜色值连续
mat = np.zeros((20,20))
for i in range(20):
    for j in range(20):
        mat[i, j] = i + j
# plt.imshow(mat)
# plt.show()
label = mat.flatten()
print(label.shape, mat.shape)

plot_eigenspace(mat, label)

# 图片
```
