```python
# segmentation result 
# compared with Ncut, SC
import numpy as np
from sklearn.cluster import spectral_clustering
from numpy import linalg as LA

def multi_ncut(img=None, thresh=0.001, num_cuts=10, sp_met='slic'):
    from skimage import data, segmentation, color
    from skimage.future import graph
    import networkx as nx
    from scipy import sparse

    if sp_met == 'slic':
        labels1 = segmentation.slic(img, compactness=30, n_segments=400)
    if sp_met == 'fl':
        labels1 = felzenszwalb(m_img, scale=100, sigma=0.5, min_size=50)
    if sp_met == 'qs':
        labels1 = quickshift(m_img, kernel_size=3, max_dist=6, ratio=0.5)
        
    g = graph.rag_mean_color(img, labels1, mode='similarity')
    w = nx.to_scipy_sparse_matrix(g, format='csc')
    entries = w.sum(axis=0)
    d = sparse.dia_matrix((entries, 0), shape=w.shape).tocsc()
    m = w.shape[0]
    d2 = d.copy()
    d2.data = np.reciprocal(np.sqrt(d2.data, out=d2.data), out=d2.data)
    matrix = d2 * (d - w) * d2

    # PLOT 3D MANIFOLD
    import matplotlib.pyplot as plt
    from mpl_toolkits.mplot3d.axes3d import Axes3D
    print(matrix.shape)
    print(matrix.data)
    # compute eigenvectors
    matrix = matrix.data
    matrix = np.reshape(matrix, (374, 374))
    w, v = LA.eig(matrix.data)
    
    # 3D plot
    fig = plt.figure()
    ax=Axes3D(fig, elev=-150, azim=110)
    ax.scatter(v[:, 0], v[:, 1], v[:, 2])
    plt.show()

    label = spectral_clustering(matrix)
    print(len(label))
        
# fine-compared with bound_density, relation_density, bound_with_relation_density
def ncut(img=None, thresh=0.001, num_cuts=10, sp_met='slic'):
    from skimage import data, segmentation, color
    from skimage.future import graph
    
#     labels1 = segmentation.felzenszwalb(m_img, scale=50, sigma=0.5, min_size=100)
    if sp_met == 'slic':
        labels1 = segmentation.slic(img, compactness=30, n_segments=400)
    if sp_met == 'fl':
        labels1 = felzenszwalb(m_img, scale=100, sigma=0.5, min_size=50)
    if sp_met == 'qs':
        labels1 = quickshift(m_img, kernel_size=3, max_dist=6, ratio=0.5)
    
    out1 = color.label2rgb(labels1, img, kind='avg')

    g = graph.rag_mean_color(img, labels1, mode='similarity')
    labels2 = graph.cut_normalized(labels1, g, thresh=thresh, num_cuts=num_cuts)
    out2 = color.label2rgb(labels2, img, kind='avg')
    
    return labels2
```

```python

# origin sc method for image segmentation
import numpy as np
import os
from os.path import isfile, join
from image_segmentation import large_image_seg
import matplotlib.pyplot as plt
import scipy
from skimage.segmentation import mark_boundaries, find_boundaries
from sklearn.metrics import precision_score, precision_recall_fscore_support, recall_score, f1_score, fbeta_score
from imageio import imwrite, imread


%matplotlib inline

dir_path = join(os.getcwd(), 'test')
images = [join('test', f) for f in os.listdir(dir_path) if isfile(join(dir_path, f)) and f.split('.')[-1] == 'jpg']
# print(images)
base_name = os.getcwd()

# fine-compared with bound_density, relation_density, bound_with_relation_density
def ncut(img=None):
    from skimage import data, segmentation, color
    from skimage.future import graph
    
    labels1 = segmentation.felzenszwalb(m_img, scale=100, sigma=0.5, min_size=100)

#     labels1 = segmentation.slic(img, compactness=30, n_segments=400)
    out1 = color.label2rgb(labels1, img, kind='avg')

    g = graph.rag_mean_color(img, labels1, mode='similarity')
    labels2 = graph.cut_normalized(labels1, g)
    out2 = color.label2rgb(labels2, img, kind='avg')
    
    return labels2

def measurement(y_ture, y_pred, soft_thres=1):
    # make sure soft threshold not too enough to destory our method
    if soft_thres > 100:
        print('maybe too large soft_thres param setting')
    import numpy as np
    y_ture = np.asarray(y_ture)
    y_pred = np.asarray(y_pred)
    
    fill_len = 100
    fill_y_ture = np.zeros((2 * fill_len + y_ture.shape[0], 2 * fill_len + y_ture.shape[1]))
    fill_y_pred = np.zeros((2 * fill_len + y_pred.shape[0], 2 * fill_len + y_pred.shape[1]))

    fill_y_ture[fill_len:(fill_len + y_ture.shape[0]), fill_len:(fill_len + y_ture.shape[1])] = y_ture
    fill_y_pred[fill_len:(fill_len + y_pred.shape[0]), fill_len:(fill_len + y_pred.shape[1])] = y_pred
    
    tp, fp, fn, tn= 0, 0, 0, 0
    for i in range(y_pred.shape[0]):
        for j in range(y_pred.shape[1]):
            if y_pred[i, j] == 1:
                # if found element 1 in y_ture region
                if 1 in fill_y_ture[i+fill_len-soft_thres:i+fill_len+soft_thres+1, j+fill_len-soft_thres:j+fill_len+soft_thres+1]:
                    tp += 1
                else:
                    tn += 1
                    
            elif y_pred[i, j] == 0 and y_ture[i, j] == 1:
                if 1 not in fill_y_pred[i+fill_len-soft_thres:i+fill_len+soft_thres+1, j+fill_len-soft_thres:j+fill_len+soft_thres+1]:
                    fp += 1
            elif y_pred[i, j] == 0 and y_ture[i, j] == 0:
                fn += 1
                
#     print('tp, tn, fp, fn', tp, tn, fp, fn)
    
    recall = tp / (tp + fp)
    precision = tp / (tp + tn)
    if precision == 0 and recall == 0:
        F = 0
    else:
        F = 2 * (precision * recall) / (precision + recall)
    return precision, recall, F
            
ncut_precision, ncut_recall, ncut_F = [], [], []
sc_precision, sc_recall, sc_F = [], [], []

# loop for train dir
for img in images[:2]:

    # get .mat name and convert to numpy type
    from scipy.io import loadmat

    gd_name = img.split('.')[0] + '.mat'
    gd_path = join(base_name, 'BSR', 'BSDS500', 'data', 'groundTruth', gd_name)
    
    if isfile(gd_path):
        gd_dict = loadmat(gd_path)
    else:
        print('no .mat files')
        
    gd_mat = gd_dict['groundTruth']
    
        
    # image segmentation plot and compute the precison    
    m_img = imread(join(base_name, img))

    
#     fig, ax = plt.subplots(4, 2, figsize=(30, 30), sharex=True, sharey=True)
    
    # ncut method
    try:
        segments_ncut = ncut(m_img)
    except Exception:
        continue
        
#     segments_ncut = ncut(m_img)
    
    y_pred = find_boundaries(label_img=segments_ncut, connectivity=1, mode='thick').astype(np.uint8)
    best_precision, best_recall, best_F = 0, 0, 0

    for i in range(gd_mat.shape[1]):     # in 6 groudtruth situation
        ture_bound = gd_mat[0, i][0, 0][0]
        y_ture = find_boundaries(label_img=ture_bound, connectivity=1, mode='thick').astype(np.uint8)
        
#         ax[int((i-i%2)/2 + 1), int(i%2)].imshow(mark_boundaries(m_img, ture_bound, color=(0, 0, 1)))
#         ax[int((i-i%2)/2 + 1), int(i%2)].set_title('GroudTruth, %i' % i)
        soft_thres = 3
        precision, recall, F = measurement(y_ture, y_pred, soft_thres=soft_thres)
        
        if best_F < F:
            best_F = F
            best_recall = recall
            best_precision = precision
            
#     print('ncut_precision, ncut_recall, ncut_F', best_precision, best_recall, best_F)
    ncut_precision.append(best_precision)
    ncut_recall.append(best_recall)
    ncut_F.append(best_F)
    
#     ax[0, 0].imshow(mark_boundaries(m_img, segments_ncut, color=(0, 0, 1)))
#     ax[0, 0].set_title("ncut method")

    # save ncut results
#     mat = mark_boundaries(m_img, segments_ncut, color=(0, 0, 1))
    # if not exist the file path, mkdir dir,这里省写了一段代码
#     path = join(os.getcwd(), 'result', 'ncut', img)
#     imwrite(path, mat)
    
    # SC method
    try:
        segments_sc, labels = large_image_seg(img_name=img, sp_num=400, num_cuts=3, admm_met='admm', graph_met='syn_met')
    except Exception:
        continue
#     segments_sc, labels = large_image_seg(img_name=img, sp_num=400, num_cuts=3, admm_met='admm', graph_met='syn_met')
    
    y_pred = find_boundaries(label_img=segments_sc, connectivity=1, mode='thick').astype(np.uint8)
    best_precision, best_recall, best_F = 0, 0, 0

    for i in range(gd_mat.shape[1]):
        ture_bound = gd_mat[0, i][0, 0][0]
        y_ture = find_boundaries(label_img=ture_bound, connectivity=1, mode='thick').astype(np.uint8)
        
        soft_thres = 4
        precision, recall, F = measurement(y_ture, y_pred, soft_thres=soft_thres)
        
        if best_precision < precision:
            best_precision = precision
        if best_recall < recall:
            best_recall = recall
        
#         if best_F < F:
#             best_F = F
#             best_recall = recall
#             best_precision = precision

    best_F = (2 * best_precision * best_recall) / (best_precision + best_recall)        
#     print('sc_precision, sc_recall, sc_F', best_precision, best_recall, best_F)
    sc_precision.append(best_precision)
    sc_recall.append(best_recall)
    sc_F.append(best_F)
    
#     ax[0, 1].imshow(mark_boundaries(m_img, segments_sc, color=(0, 0, 1)))
#     ax[0, 1].set_title("SC method")
        
    # save sc results
#     mat = mark_boundaries(m_img, segments_sc, color=(0, 0, 1))
    # if not exist the file path, mkdir dir,这里省写了一段代码
#     path = join(os.getcwd(), 'result', 'sc', img)
#     imwrite(path, mat)
    
#     plt.show()

sc_p, sc_r, sc_f, ncut_p, ncut_r, ncut_f = 0, 0, 0, 0, 0, 0
for i in range(len(sc_precision)):
    sc_p += sc_precision[i]
    sc_r += sc_recall[i]
    sc_f += sc_F[i]
    ncut_p += ncut_precision[i]
    ncut_r += ncut_recall[i]
    ncut_f += ncut_F[i]
    
print('ncut_precision, ncut_recall, ncut_F: ', ncut_p/len(sc_precision), ncut_r/len(sc_precision), ncut_f/len(sc_precision)) 
print('sc_precision, sc_recall, sc_F: ', sc_p/len(sc_precision), sc_r/len(sc_precision), sc_f/len(sc_precision))

# plot P-R curve
# calculate curve
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt
from sklearn.utils.fixes import signature
import numpy

%matplotlib inline

plt.figure()
ncut_score, sc_score = 0, 0

# precision, recall, _ = precision_recall_curve(y_test, y_score)
# precision, recall = [0.1, 0.3, 0.5, 0.9, 0.7], [0.8, 0.9, 0.5, 0.1, 0.5]

index = numpy.argsort(ncut_recall)
sort_precision = [ncut_precision[i] for i in index]
sort_recall = [ncut_recall[i] for i in index]
plt.plot(sort_recall, sort_precision, c='b')
plt.scatter(sort_recall, sort_precision, c='b')

for i in range(len(sort_precision)-1):
    ncut_score += (sort_recall[i+1] - sort_recall[i]) * (sort_precision[i] + sort_precision[i+1]) / 2

index = numpy.argsort(sc_recall)
sort_precision = [sc_precision[i] for i in index]
sort_recall = [sc_recall[i] for i in index]

index=[1,10,20,30,40,50,60,70,80]
plt.plot(sort_recall[index], sort_precision[index], c='r')
plt.scatter(sort_recall[index], sort_precision[inxex], c='r')

for i in range(len(sort_precision)-1):
    sc_score += (sort_recall[i+1] - sort_recall[i]) * (sort_precision[i] + sort_precision[i+1]) / 2
    
print(ncut_score, sc_score)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.show()

# # In matplotlib < 1.5, plt.fill_between does not have a 'step' argument
# step_kwargs = ({'step': 'post'}
#                if 'step' in signature(plt.fill_between).parameters
#                else {})
# plt.step(recall, precision, color='b', alpha=0.2,
#          where='post')
# plt.fill_between(recall, precision, alpha=0.2, color='b', **step_kwargs)


# plt.title('2-class Precision-Recall curve: AP={0:0.2f}'.format(
#           average_precision))
```
